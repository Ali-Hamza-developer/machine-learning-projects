## Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

## Model Training

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

## Model Evaluation 

## Prediction on training data

```python
print(training_data_prediction)
```

### R Square Error

```python
from sklearn import metrics
```

```python
score_1 = metrics.r2_score(Y_train, training_data_prediction)
```

```python
print("R Square Error:", score_1)
```

## Mean Absolute Error

```python
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
```

```python
print("Mean Absolute Error:", score_2)
```

### Why We Use RÂ² and Mean Absolute Error in House Price Prediction

In regression problems like **house price prediction**, we aim to estimate a **continuous value** (the price) rather than classify data into categories. This is why **accuracy**, which measures the fraction of exact correct predictions, is **not suitable** for regression tasks.

Instead, we use metrics that capture how close our predicted prices are to the actual values:

1. **R-squared (RÂ²) Score**
   - Indicates how well the model explains the variability of the target variable.
   - Ranges from 0 to 1 (or can be negative if the model is very poor).
   - Higher RÂ² means the model predictions fit the actual data better.
   

2. **Mean Absolute Error (MAE)**
   - Measures the **average absolute difference** between predicted and actual values.
   - Provides an intuitive understanding of **how much predictions deviate** from true prices.
   - Lower MAE indicates better prediction performance.

**Key Takeaway:**  
Regression models need metrics that reflect **continuous error**, not just exact matches. Thatâ€™s why RÂ² and MAE are preferred over accuracy.
