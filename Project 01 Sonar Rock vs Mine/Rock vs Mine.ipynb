## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Preprocessing 

```python
df = pd.read_csv("Copy of sonar data.csv", header=None)
```

```python
df.head(5)
```

```python
#number of rows and columns
df.shape
```

```python
#statistical Analysis
df.describe()
```

### Class Distribution (Rock vs Mine)

The dataset contains two classes:
- **Rock (R):** 97 samples  
- **Mine (M):** 111 samples  




```python
df[60].value_counts()
```

```python
df.groupby(60).mean()
```

### Separating Features and Labels

The dataset is divided into:
- **Features (X):** All input variables used for prediction  
- **Labels (y):** Target variable indicating Rock or Mine


```python
X = df.drop(columns=60, axis=1)
Y = df[60]
```

```python
print(X)
print(Y)
```

## Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y , random_state=1)
```

```python
print(X.shape)   #orignal dataset
print(X_train.shape) # Train data
print(X_test.shape)  # Testing Data
```

## Model Training ---> Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

#### Training Logistic Regression with Training data

```python
model.fit(X_train, Y_train)
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
# Accuracy on Training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training Dataset :" ,training_data_accuracy)
```

```python
# Accuracy on Test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test Dataset :" ,test_data_accuracy)
```
